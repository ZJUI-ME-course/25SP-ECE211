```python
import numpy as np
import matplotlib.pyplot as plt

# 定义频率范围
omega = np.linspace(0, 20, 100)  # 从0到20弧度/秒，1000个点
# print(omega)
# 定义传递函数的分子和分母
numerator = 1j * omega - 2 * omega**2
denominator = 4 - omega**2 + 1j * 4 * omega

# 计算传递函数的值
H = numerator / denominator
print(H)

# 幅度响应
magnitude = np.abs(H)
print(magnitude)

# 相位响应
phase = np.angle(H, deg=True)  # 将相位转换为度
print(phase)


# 绘制幅度响应
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(omega, magnitude, label=r'$|H(\omega)|$')
plt.xlabel(r'$\omega$ (rad/s)')
plt.ylabel(r'$|H(\omega)|$')
plt.title(r'Magnitude Response of $H(\omega)$')
plt.grid(True)
plt.legend()

# 绘制相位响应
plt.subplot(2, 1, 2)
plt.plot(omega, phase, label=r'$\angle H(\omega)$', color='orange')
plt.xlabel(r'$\omega$ (rad/s)')
plt.ylabel(r'$\angle H(\omega)$ (degrees)')
plt.title(r'Phase Response of $H(\omega)$')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
