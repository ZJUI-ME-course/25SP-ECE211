```python
import numpy as np
import matplotlib.pyplot as plt

# 
omega = np.linspace(0.01, 90, 10000)  # 020/1000
# print(omega)
# 
numerator = 1
denominator = 1+1j*(omega*0.2-1/omega/0.05)

# 
H = numerator / denominator
# print(H)

# 
magnitude = np.abs(H)
# print(magnitude)

# 
phase = np.angle(H, deg=True)  # 
# print(phase)


# 
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
plt.plot(omega, magnitude, label=r'$|H(\omega)|$')
plt.xlabel(r'$\omega$ (rad/s)')
plt.ylabel(r'$|H(\omega)|$')
plt.title(r'$|H(\omega)|$')

# plt.title(r'R=10')
plt.grid(True)
plt.legend()

# 
plt.subplot(2, 1, 2)
plt.plot(omega, phase, label=r'$\angle H(\omega)$', color='orange')
plt.xlabel(r'$\omega$ (rad/s)')
plt.ylabel(r'$\angle H(\omega)$ (degrees)')
plt.title(r'Phase Response of $H(\omega)$')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
