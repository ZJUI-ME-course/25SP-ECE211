```python
import numpy as np
import matplotlib.pyplot as plt

# 
omega = np.linspace(0.01, 90, 10000)  # 020/1000
# print(omega)
# 
numerator = 1
denominator = 1+1j*(omega*0.2-1/omega/0.05)

# 
H = numerator / denominator
# print(H)

# 
magnitude = np.abs(H)
# print(magnitude)

# 
phase = np.angle(H, deg=True)  # 
# print(phase)


# 
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
plt.plot(omega, magnitude, label=r'$|H(\omega)|$')
plt.xlabel(r'$\omega$ (rad/s)')
plt.ylabel(r'$|H(\omega)|$')
plt.title(r'$|H(\omega)|$')

# plt.title(r'R=10')
plt.grid(True)
plt.legend()

# 
plt.subplot(2, 1, 2)
plt.plot(omega, phase, label=r'$\angle H(\omega)$', color='orange')
plt.xlabel(r'$\omega$ (rad/s)')
plt.ylabel(r'$\angle H(\omega)$ (degrees)')
plt.title(r'Phase Response of $H(\omega)$')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# ft

import numpy as np
import matplotlib.pyplot as plt

# 
t = np.linspace(0, 4*np.pi, 1000)

# (a) f(t) = 1/2 + 1/2 cos(2t)
f_a = 0.5 + 0.5 * np.cos(2 * t)

# (b) f(t) = 2 + 2j sin(3t)
f_b = 2 + 2 * np.sin(3 * t)

# (c) f(t) = 2 cos(2t) + 3 cos(3t)
f_c = 2 * np.cos(2 * t) + 3 * np.cos(3 * t)

# (d) f(t) = -2j sin(2t) + 2 cos(4t)
f_d = -2 * np.cos(2 * t) + 2 * np.cos(4 * t)

# 
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(t, f_a)
plt.title('(a) f(t) = 1/2 + 1/2 cos(2t)')
plt.xlabel('Time (s)')
plt.ylabel('f(t)')
plt.xlim(0,2*np.pi)

plt.subplot(2, 2, 2)
plt.plot(t, f_b)
plt.title('(b) f(t) = 2 + 2 sin(3t)')
plt.xlabel('Time (s)')
plt.ylabel('f(t)')
plt.xlim(0,4/3*pi)

plt.subplot(2, 2, 3)
plt.plot(t, f_c)
plt.title('(c) f(t) = 2 cos(2t) + 3 cos(3t)')
plt.xlabel('Time (s)')
plt.ylabel('f(t)')
plt.xlim(0,4*pi)


plt.subplot(2, 2, 4)
plt.plot(t, f_d)
plt.title('(d) f(t) = -2 cos(2t) + 2 cos(4t)')
plt.xlabel('Time (s)')
plt.ylabel('f(t)')
plt.xlim(0,2*pi)



plt.tight_layout()
plt.show()
```
