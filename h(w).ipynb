```python
import numpy as np
import matplotlib.pyplot as plt

# 定义频率范围
omega = np.linspace(0.01, 90, 10000)  # 从0到20弧度/秒，1000个点
# print(omega)
# 定义传递函数的分子和分母
numerator = 1
denominator = 1+1j*(omega*0.2-1/omega/0.05)

# 计算传递函数的值
H = numerator / denominator
# print(H)

# 幅度响应
magnitude = np.abs(H)
# print(magnitude)

# 相位响应
phase = np.angle(H, deg=True)  # 将相位转换为度
# print(phase)


# 绘制幅度响应
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
plt.plot(omega, magnitude, label=r'$|H(\omega)|$')
plt.xlabel(r'$\omega$ (rad/s)')
plt.ylabel(r'$|H(\omega)|$')
plt.title(r'$|H(\omega)|$')

# plt.title(r'R=10')
plt.grid(True)
plt.legend()

# 绘制相位响应
plt.subplot(2, 1, 2)
plt.plot(omega, phase, label=r'$\angle H(\omega)$', color='orange')
plt.xlabel(r'$\omega$ (rad/s)')
plt.ylabel(r'$\angle H(\omega)$ (degrees)')
plt.title(r'Phase Response of $H(\omega)$')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义频率范围
omega = np.linspace(0.1, 90, 10000)  # 避免除以零，从0.1开始

# 定义传递函数的幅度
H = 1 / np.sqrt(1 + (0.2 * omega**21 - 20 / omega)**2)

# 绘制幅度响应
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
plt.plot(omega, H, label=r'$|H(\omega)|$')  # 修正变量名
plt.xlabel(r'$\omega$ (rad/s)')
plt.ylabel(r'$|H(\omega)|$')
plt.title(r'Amplitude Response $|H(\omega)|$')
plt.grid(True)
plt.legend()

# 如果需要绘制相位响应，可以继续添加代码
# 这里只绘制幅度响应
plt.show()
```
