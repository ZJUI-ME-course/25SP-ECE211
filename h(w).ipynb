```python
import numpy as np
import matplotlib.pyplot as plt

# 
omega = np.linspace(0, 20, 100)  # 020/1000
# print(omega)
# 
numerator = 1j * omega - 2 * omega**2
denominator = 4 - omega**2 + 1j * 4 * omega

# 
H = numerator / denominator
print(H)

# 
magnitude = np.abs(H)
print(magnitude)

# 
phase = np.angle(H, deg=True)  # 
print(phase)


# 
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(omega, magnitude, label=r'$|H(\omega)|$')
plt.xlabel(r'$\omega$ (rad/s)')
plt.ylabel(r'$|H(\omega)|$')
plt.title(r'Magnitude Response of $H(\omega)$')
plt.grid(True)
plt.legend()

# 
plt.subplot(2, 1, 2)
plt.plot(omega, phase, label=r'$\angle H(\omega)$', color='orange')
plt.xlabel(r'$\omega$ (rad/s)')
plt.ylabel(r'$\angle H(\omega)$ (degrees)')
plt.title(r'Phase Response of $H(\omega)$')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
